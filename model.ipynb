```python
import numpy as np
import pandas as pd
import string
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("Tweets.csv")
df.head()
df1=df.copy()
df1=df1.drop(columns=["textID","selected_text"])
df1.text=df1.text.fillna("")

df1.text=df1.text.apply(lambda x:x.lower())
def rem_num(txt):
    a=""
    for i in txt:
        if not i.isdigit():
            a+=i
    return a

def rem_punctuation(txt):
    return txt.translate(txt.maketrans("","",string.punctuation))

def rem_emoji(text):
    a=""
    for i in text:
        if i.isascii():
            a+=i
    return a

def rem_link(txt):
    return re.sub(r'http\S+|www\S+|https\S+', '', txt)

stop_words=set(stopwords.words("english"))


def rem_stopwords(txt):
    tokenize=word_tokenize(txt)
    return " ".join([words for words in tokenize if words not in stop_words])

def clean_data(txt):
    txt=rem_num(txt)
    txt=rem_punctuation(txt)
    txt=rem_emoji(txt)
    txt=rem_link(txt)
    txt=rem_stopwords(txt)
    return txt

   
df1["text"]=df1["text"].apply(clean_data)

df1.head()



```

```python
map={}
count=1
for i in df1.sentiment.unique():
    if i not in map:
        map[i]=count
        count+=1
print(map)
df1.sentiment=df1.sentiment.map(map)
df1.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df1.text, df1.sentiment, test_size=0.20, random_state=42,stratify=df1.sentiment)

bow_vectorizer=CountVectorizer(ngram_range=(1,2),)
bow_X_train=bow_vectorizer.fit_transform(X_train)
bow_X_test=bow_vectorizer.transform(X_test)

model=LogisticRegression()
model.fit(bow_X_train,y_train)
bow_y_pred=model.predict(bow_X_test)
print(classification_report(y_test,bow_y_pred))

a=input("enter the text")
bow_X=bow_vectorizer.transform([a])
bow_y=model.predict(bow_X)

if bow_y[0]==1:
    print("neutral")
elif bow_y[0]==2:
    print("negative")
else:
    print("positive")



```

```python
tfidf_vectorizer=TfidfVectorizer(ngram_range=(1,2))
tfidf_X_train=tfidf_vectorizer.fit_transform(X_train)
tfidf_X_test=tfidf_vectorizer.transform(X_test)

model=LogisticRegression()
model.fit(tfidf_X_train,y_train)
tfidf_y_pred=model.predict(tfidf_X_test)
print(classification_report(y_test,tfidf_y_pred))

a=input("enter the text")
bow_X=bow_vectorizer.transform([a])
bow_y=model.predict(bow_X)

if bow_y[0]==1:
    print("neutral")
elif bow_y[0]==2:
    print("negative")
else:
    print("positive")
```
